```python
#CDF2
def cum_sum(data):
    lst1 = np.sort(data)
    ew_lst = []
    for i in lst1:
        if i not in ew_lst:
            ew_lst += [i]
    new_lst = []
    count = 0
    for j in ew_lst:
        for k in data:
            if j == k:
                count += 1
        new_lst += [count]
    new1 = []
    for i in new_lst:
        cum_prob = i / len(data)
        new1 += [cum_prob]
    return ew_lst, new1
    
    
data = df['Price']
x, y = cum_sum(data)
```

```python
plt.title('CDF for Price')
plt.xlabel('Price')
plt.ylabel('Cumulative Probability')
plt.grid()
plt.plot(x, y)
plt.show()
plt.savefig("plot.png")
plt.show()
```

```python
import numpy as np
def CI(data, size, sample):
    lst1 = []
    for i in range(1, sample + 1):
        sample1 = np.random.choice(data, size, replace = True)
        lst1.append(np.median(sample1))
    sorts = np.sort(lst1)
    a = np.percentile(sorts, 95)
    b = np.percentile(sorts, 1)
    return [b, a]
data = df['Price']
size = 30
sample = 1000
CI(data, size, sample) 
```
