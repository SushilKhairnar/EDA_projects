```python
# CDF
def cum_sum(data):
    lst1 = np.sort(data)
    ew_lst = []
    for i in lst1:
        if i not in ew_lst:
            ew_lst += [i]
    new_lst = []
    count = 0
    for j in ew_lst:
        for k in data:
            if j == k:
                count += 1
        new_lst += [count]
    new1 = []
    for i in new_lst:
        cum_prob = i / len(data)
        new1 += [cum_prob]
    return ew_lst, new1
    
    
data = np.log(df['ApplicantIncome'])
x, y = cum_sum(data)
```

```python
# it is mototonically increasing graph and also it is looking like a somewhat gaussian 
plt.title('CDF for ApplicantIncome')
plt.xlabel('ApplicantIncome')
plt.ylabel('Cumulative Probability')
plt.grid()
plt.plot(x, y)
plt.show()
```

```python
sample_means = []
def apply_clt(data, size_of_sample, no_of_samples):
    for i in range(0, no_of_samples):
        sample = np.random.choice(data, size = size_of_sample, replace = True)
        sample_means.append(np.mean(sample))
size_of_sample = 30
no_of_samples = 1000
data = df['ApplicantIncome'][df['Gender'] == 'Female']
apply_clt(data, size_of_sample, no_of_samples)
```

```python
# the below graph somewhat looking like a gaussian distribution 
# all the points is symmetric towards the mean
# most of the female income is lies between 4000 RS to 4800 RS
sns.distplot(sample_means)
plt.xlabel('sample means')
plt.ylabel('female_income')
plt.title('CLT on female income')
plt.grid()
plt.show()
```
